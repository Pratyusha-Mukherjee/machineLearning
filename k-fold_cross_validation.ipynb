```python
data = load_iris()
X, y = data.data, data.target
```

```python


clf = DecisionTreeClassifier()


scores = cross_val_score(clf, X, y, cv=10)


print("Cross-validation scores:\n", scores)
print("Mean accuracy:\n", scores.mean())
print("Standard deviation:\n", scores.std())


plt.figure(figsize=(10, 6))
plt.boxplot(scores, vert=False, patch_artist=True)
plt.title('10-Fold Cross-Validation Scores of Decision Tree Classifier')
plt.xlabel('Accuracy')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

# Load dataset (using Iris dataset for this example)
data = load_iris()
X, y = data.data, data.target

# Initialize the Decision Tree classifier
clf = DecisionTreeClassifier()

# Perform 5-fold cross-validation
scores = cross_val_score(clf, X, y, cv=5)

# Print the cross-validation scores
print("Cross-validation scores:", scores)
print("Mean accuracy:", scores.mean())
print("Standard deviation:", scores.std())

# Plot the results using a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(range(1, len(scores) + 1), scores, color='blue', label='Fold Accuracy')
plt.axhline(y=scores.mean(), color='red', linestyle='--', label='Mean Accuracy')
plt.title('5-Fold Cross-Validation Scores of Decision Tree Classifier')
plt.xlabel('Fold')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
